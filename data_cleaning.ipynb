```python
import pandas as pd
import numpy as np
```

```python
# Import income data
income_df = pd.read_csv("income_evaluation.csv")
income_df[10:15]
```

```python
income_df.columns
```

```python
# Remove the blank from column name
income_df.columns = income_df.columns.str.replace(' ', '')
income_df.columns
```

```python
# Replace '?' to NAN
income_df[['workclass','occupation','native-country']] = income_df[['workclass','occupation','native-country']].replace(' ?',np.nan)
income_df[10:15]
```

```python
# Find NAN data
income_df.isnull().sum()
```

```python
income_df.count()
```

```python
# Drop NAN data raw
income_df = income_df.dropna()
income_df.count()
```

```python
income_df.dtypes
```

```python
# Drop unnecessary columns
income_df = income_df.drop(["fnlwgt","education-num","capital-gain","capital-loss"],axis=1)
income_df.head(10)
```

```python
income_df.groupby(['income']).count()['age']
```

```python
# Save to csv
income_df.to_csv("data_cleaning_cleaned.csv", index=False)
```
